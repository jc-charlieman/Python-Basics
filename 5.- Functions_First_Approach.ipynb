### print() Function

Displays messages or variable values to the user.
Syntax: print(Object(s), sep=, end=)

```python
# Al colocar la función print sin argumento arroja el salto de línea
# print function without argument displays a new line
print()
```

```python
# print function with explicit new line (the same result as previous)
print(end='\n')
```

```python
# print function with implicit arguments (sep='' and end='\n')
print('Welcome!','Pythonista')
print('Hey')
```

```python
# print function with explicit argument (same result as previous)
print('Hi!','Pythonista', sep='---',end='\n')   
print('Hey')
```

```python
# print function with string, tuple and list as arguments
print('Hello!',('Hi',8,7.9),[1,2,3], sep='--')
```

### input() function

Ask the user to introduce an answer. Syntax is the following:
_input('Message')_.
Read values are considered as strings by default. It is possible to use int, float, tuple ans list functions to modify the original values.

```python
input('type a number: ')
```

```python
var = input("How old are you?: ")
print(var)
```

```python
print(type(var))
```

```python
vari = int(input('type the current year: '))
print(type(vari))
```

```python
varf = float(input('type a number: '))
print(type(varf))
```

```python
vart = tuple(input('type a number: '))
print(vart)
print(type(vart))
```

```python
varl = list(input('type a number: '))
print(varl)
print(type(varl))
```

```python
# input function reading several variables
v_0, v_1 = input('type 2 words: ').split('*')
print(v_0,type(v_0))
print(v_1,type(v_1))
```

```python
variable_1, variable_2 = 15, 30
print(variable_1, variable_2)
```

```python
# input function with variable
name = 'Pepe'
ln = input(f'What"s your last name {name}:')
print('Welcome',name,ln,'.')
```
