## if - elif - else:

```python
# simple if structure
age = 21 
if age >= 18:
    print("¡You're of age!")
```

```python
# if & else 
age = 15
if age >= 18:
    print("You're of age!")
else:
    print("You're a minor!")
```

```python
# if & else 
n = int(input('Type a number: '))      
if n%2 == 0:
    print('The number {} is even'.format(n))
else:
    print(f'The number {n} is odd.')
```

```python
# if - elif - else 
total = float(input('Type the total amount: '))
if total <= 1000:
    print('Cash payment.')
elif 1000  < total < 5000:
    print('Debit card payment.')
else:
    print('Credit card payment.')
```

```python
# if-elif-else API Classification 
api = float(input('Type oil specific gravity: '))
if api > 31.1:
    print('Light oil')
elif 22.3 <= api <= 31.1:
    print('Medium oil')
elif 10 <= api < 22.3:
    print('Heavy oil')
else:
    print('Extra Heavy oil')
```

```python
# Nested if-elif-else Mexican crude oil classification
API = float(input('Type oil specific gravity: '))
if API > 31.1:
    if API == 39:
        print('Olmeca oil')
    elif API == 32:
        print('Istmo oil')
    else:
        print('Light oil')
elif 22.3 <= API <= 31.1:
    print('Medium oil')
elif 10 <= API < 22.3:
    if API == 22:
        print('Maya oil')
    else:
        print('Heavy oil')
elif API < 10:
    print('Extraheavy oil')
```

## Maximum & minimum values

```python
# Maximum value
A = [7,2,24,18,34,5]
Max = A[0]
for i in A:
    if i > Max:
        Max = i
print(Max)
```

```python
# Minimum value
A = [7,2,24,18,34,5]
Min = A[0]
for i in A:
    if i < Min:
        Min = i
print(Min)
```

```python
A = [7.5,2.4,24,18,34.8,5]
maximum = max(A)
print(maximum)
```

```python
A= [7,2,24,18,34,5]
minimum = min(A)
print(minimum)
```

### Match Case Statement

```python
value = int(input('Type an integer between 10 and 50:'))
match value:
    case value if 0 < value <= 10:
        print('Number between 0-10.')
    case value if 10 < value <= 20:
        print('Number between 10-20.')
    case value if 20 < value <= 30:
        print('Number between 20-30.')
    case value if 30 < value <= 40:
        print('Number between 30-40.')
    case value if 40 < value <= 50:
        print('Number between 40-50.')
    case _:
        print('Number out of range.')
```
